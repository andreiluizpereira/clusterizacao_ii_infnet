### Faixa dinâmica e preparação para clusterização

Foram gerados boxplots para todas as variáveis socioeconômicas e de saúde, mostrando grande variação de escala entre elas.
Por exemplo, o PIB per capita `(gdpp)` apresenta valores muito maiores do que a taxa de mortalidade infantil `(child_mort)`.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.boxplot(data=df.drop('country', axis=1))
plt.xticks(rotation=45)
plt.title('Distribuição (faixa dinâmica) das variáveis')
plt.show()
```
